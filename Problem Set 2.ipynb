```python
def plot_fill(myexpr,myvar,myrange,myrange_fill):
    myx=np.linspace(myrange[0],myrange[1],1000)
    myx1=np.linspace(myrange_fill[0],myrange_fill[1],1000)
    lam_function=lambdify(myvar,myexpr,'numpy')
    plt.plot(myx,lam_function(myx))
    plt.fill_between(myx1,lam_function(myx1),color='grey')
    plt.show()
```
